This is a Word2Vec example using CBOW (Continuous Bag Of Words) and Skip-Gram model to generate word embeddings. 

```python
# if not already installed in your environment install gensim librarty to access gensim model Word2Vec
#!pip install gensim
```

```python
# ===========================================
# Import required libraries
# ===========================================
import nltk
import gensim
from gensim.models import Word2Vec # for learning word embeddings from large amounts of text data
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt')
import warnings

warnings.filterwarnings(action='ignore')
```

```python
# -------------------------------------------
# Load corpora
# -------------------------------------------

with open("smartphone_corpus_100k.txt", "r", encoding="utf-8") as f:
    smartphone_corpus = f.read().lower()

# Tokenize the text
data = []

for i in sent_tokenize(smartphone_corpus):
    sent = []

    for j in word_tokenize(i):
        sent.append(j.lower())

    data.append(sent)
```

```python
# -------------------------------------------
# Train Word2Vec Models - CBOW & Skip-Gram
# -------------------------------------------

print("Training Word2Vec using CBOW model")
model_cbow = gensim.models.Word2Vec(data, min_count=1,
                                vector_size=100, window=5)

print("Training Word2Vec using skip-gram model")
model_skip = gensim.models.Word2Vec(data, min_count=1, vector_size=100,
                                window=5, sg=1)

```

```python
# -------------------------------------------
# Evaluating word similarities
# -------------------------------------------

print("Cosine similarity between 'efficient' " +
      "and 'fast' - CBOW : ",
      model_cbow.wv.similarity('efficient', 'fast'))

print("Cosine similarity between 'efficient' " +
      "and 'fast' - skip gram : ",
      model_skip.wv.similarity('efficient', 'fast'))

print("Cosine similarity between 'camera' " +
      "and 'photography' - CBOW : ",
      model_cbow.wv.similarity('camera', 'photography'))

print("Cosine similarity between 'camera' " +
      "and 'photography' - skip gram : ",
      model_skip.wv.similarity('camera', 'photography'))

print("Cosine similarity between 'performance' " +
      "and 'productivity' - CBOW : ",
      model_cbow.wv.similarity('performance', 'productivity'))

print("Cosine similarity between 'performance' " +
      "and 'productivity' - skip gram : ",
      model_skip.wv.similarity('performance', 'productivity'))
```
